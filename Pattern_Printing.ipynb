1. Write code to print a Square pattern

```python
n=int(input("Enter the value of n :"))
for i in range(n):
  for j in range(n):
    print(f"*",end=" ")
  print()
```

2.Write code to print Right Triangle pattern

```python
n=int(input("Enter the value of n :"))
for i in range(1,n+1):
  for j in range(i):
    print(f"*",end=" ")
  print()
```

3. Write code to print Inverted Right Triangle

```python
n=int(input("Enter the value of n :"))
for i in range(n,0,-1):
  for j in range(i):
    print(f"*",end=" ")
  print()
```

4. Write code to print Number Triangle pattern

```python
n=int(input("Enter the value of n :"))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end=" ")
  print()
```

5. Write code to print pyramid pattern

```python
n=int(input("Enter the value of n :"))
for i in range(1,n+1):
    print(f" "*(n-i),end="  ")
    print("*"*(2*i-1))
```

6. Write code to print reverse number pattern

```python
n=int(input("Enter the value of n :"))
for i in range(n,0,-1):
  for j in range(1,i+1):
    print(j,end=" ")
  print()
```

7.  Take a string from user and print it in cross of both diagonal pattern
and remaining blank space print *


```python
name=input("Enter a string :")
n=len(name)
for i in range(n):
  for j in range(n):
    if i==j and i+j==n-1:
      print(name[i],end=" ")
    elif i==j:
      print(name[i],end=" ")
    elif i+j==n-1:
      print(name[i],end=" ")
    else:
      print("*",end=" ")
  print()
```

8. Take a String from user and print it on the main diagonal of a square pattern and remaining blank space print *

```python
name=input("Enter a string :")
n=len(name)
for i in range(n):
  print(f"*"*i+name[i]+"*"* (n-i-1))
```

9. Print this pattern

```python
n=int(input("Enter a value of n :"))
for row in range(1,n+1):
  for col in range(1,n+1):
    print(max(row-1,n-row,col-1,n-col)+1,end=" ")
  print()
```

```python
n=int(input("Enter a value of n :"))
for i in range(1,n+1):
  for j in range(1,i+1):
    if j%2 !=0:
      print(f"*",end=" ")
    else:
      print(j,end=" ")
  print()
```
