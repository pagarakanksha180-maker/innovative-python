1. Write a python program to build an ola/uber Auto meter
for first 100 km , charges will be rs 10/km
for next 100 km,charges will be rs 9/km
for next 500 km,charges will be rs 6/km
for waiting time , rs 5/min




```python
distance=float(input("Enter distance from 0 travelled in km:"))
wait_time=int(input("Enter waiting time in minutes:"))
total_charge=0
n=abs(distance)
if n<=100:
  total_charge=n*10
elif n<=200:
  total_charge=(100*10)+((n-100)*9)
elif n<=700:
  total_charge=(100*10)+(100*9)+((n-200)*6)
else:
  total_charge=(100*10)+(100*9)+(500*6)+((n-700)*6)
  total_charge=total_charge+(wait_time*5)
print(f"Total_charge {total_charge}")
```

2. Write a python program to generate an Electricity Bill.
If units < 100, then bill is 'Free'
For next 100 units, then bill is Rs 5/unit
for next 100 units, then bill is Rs 7/unit
Thereafter any number of units Rs 10/unit

```python
units=int(input("Enter electricity units consumed:"))
if units<100:
  bill=0
elif units<=200:
  bill=(units-100)*5
elif units<=300:
  bill=(100*5)+(units-200)*7
else:
  bill=(100*5)+(100*7)+(units-300)*10
print("Electricity Bill=Rs",bill)
```

9.write a program to reverse a number

```python
num=int(input("Enter a number"))
negative=False
if num<0:
  negative=True
  num=-num
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
if negative:
  rev=-rev
  print(f"Reversed number:{rev}")

```

10.Write a program to count the number of digits in a given integer using a while loop

```python
n=int(input("Enter a number"))
count=0
if n==0:
  count=1
else:
  if n<0:
    n=-n
  while n>0:
    count+=1
    n=n//10
print(f"Number of digits: {count}")
```

11.Find LCM of given numbers without using inbuilt functions

```python
a=int(input("Enter first number"))
b=int(input("Enter second number"))
if a>b:
  print("a is greater")
  max_num=a
elif a==b:
  print("a and b are equal")
  max_num=a
else:
  print("b is greater")
  max_num=b
  while True:
    if max_num % a == 0 and max_num % b == 0:
      lcm = max_num
      break
    max_num += 1
  print(f"The LCM of {a}and{b} is {lcm}")
```

12. Find HCF of given numbers

```python
a=int(input("Enter first number :"))
b=int(input("Enter second number :"))
if a<b:
  small=a
else:
  small=b
hcf=1
i=1
while i<=small:
  if a%i==0 and b%i==0:
    hcf=i
  i+=1
print(f"HCF of {a} and {b} is {hcf}")

```

13.to convert an integer into binary form without using inbuilt functions

```python
num=int(input("Enter an number :"))

if num==0:
  binary_num=0
else:
  binary_num=0
  place=1
  a=num
while a>0:
  remainder=a%2
  binary_num=binary_num+(remainder*place)
  place=place*10
  a=a//2
print("Binary number=",binary_num)
```

14.Write a program to take 2 integer as a input from user start and end then print every alternate number between them

```python
start=int(input("Enter start number :"))
end=int(input("Enter end number :"))
if start>=end:
  large=start
  while large>=end:
    print(large,end=" ")
    large-=2

else:
  large=start
  while large<=end:
    print(large,end=" ")
    large +=2
```

15.Take 2 integer from user print every alternate number between them in single line print like 1_3_5_7_9
_ represents blank space dont print _ after 9

```python
start = int(input("Enter start number :"))
end = int(input("Enter end number :"))

numbers_to_print = []

if start >= end:
  current = start
  while current >= end:
    numbers_to_print.append(str(current))
    current -= 2
else:
  current = start
  while current <= end:
    numbers_to_print.append(str(current))
    current += 2

print("_".join(numbers_to_print))
```

17. to convert the string in upper case letters without using inbuilt functions

```python
s=input("Enter a string:")
result=""
for ch in s:
  if 'a'<=ch <= 'z':
    result+= chr(ord(ch)-32)
  else:
    result +=ch
print(f"Uppercase string : {result}")
```

18. to convert the string in lower case letters without using inbuilt functions

```python
s=input("Enter a string:")
result=""
for ch in s:
  if 'A'<=ch <= 'Z':
    result+= chr(ord(ch)+32)
  else:
    result +=ch
print(f"Lowercase string : {result}")
```

19. to concat two strings without using inbuilt functions

20.program to replace a string

```python
s=input("Enter first string :")
old=input("Enter character to replace:")
new=input("Enter character to replace with:")
result=""
for ch in s:
  if ch==old:
    result += new
  else:
    result += ch
print(f"New string : {result}")
```
